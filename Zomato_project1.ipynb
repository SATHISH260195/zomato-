```python
from faker import Faker
from faker.providers import BaseProvider

class CuisineProvider(BaseProvider):
    """Custom provider to generate random Indian and Chinese cuisine types."""

    cuisines = {
        "Indian": [
            "North Indian", "South Indian", "Mughlai", "Punjabi", "Gujarati",
            "Rajasthani", "Bengali", "Goan", "Maharashtrian", "Chettinad"
        ],
        "Chinese": [
            "Sichuan", "Cantonese", "Hunan", "Shanghainese", "Fujian",
            "Mandarin", "Xinjiang", "Teochew", "Yunnan", "Tibetan"
        ]
    }

    def cuisine_type(self, category=None):
        """Returns a random cuisine type from Indian or Chinese categories."""
        if category and category in self.cuisines:
            return f"{category}: {self.random_element(self.cuisines[category])}"
        else:
            random_category = self.random_element(list(self.cuisines.keys()))
            return f"{random_category}: {self.random_element(self.cuisines[random_category])}"


test = Faker()
test.add_provider(CuisineProvider)


print(test.cuisine_type("Indian"))
print(test.cuisine_type("Chinese"))
print(test.cuisine_type())

for _ in range(5):
    print(test.cuisine_type())
```

```python
import uuid
from datetime import timedelta
class Order:
    def __init__(self, order_id, customer_id, restaurant_id, order_date, delivery_time, status,
                 total_amount, payment_mode, discount_applied, feedback_rating):
        self.order_id = order_id
        self.customer_id = customer_id
        self.restaurant_id = restaurant_id
        self.order_date = order_date
        self.delivery_time = delivery_time
        self.status = status
        self.total_amount = total_amount
        self.payment_mode = payment_mode
        self.discount_applied = discount_applied
        self.feedback_rating = feedback_rating
    def __str__(self):
        return (f"Order ID: {self.order_id}\n"
                f"Customer ID: {self.customer_id}\n"
                f"Restaurant ID: {self.restaurant_id}\n"
                f"Order Date: {self.order_date}\n"
                f"Delivery Time: {self.delivery_time}\n"
                f"Status: {self.status}\n"
                f"Total Amount: ${self.total_amount}\n"
                f"Payment Mode: {self.payment_mode}\n"
                f"Discount Applied: ${self.discount_applied}\n"
                f"Feedback Rating: {self.feedback_rating if self.feedback_rating else 'N/A'}\n")
    def generate_random_order(self):
        self.order_id = str(uuid.uuid4())
        self.customer_id = str(uuid.uuid4())
        self.restaurant_id = str(uuid.uuid4())
        self.order_date = fake.date_time_between(start_date="-30d", end_date="now")
        self.delivery_time = self.order_date + timedelta(minutes=random.randint(20, 90))
        self.status = random.choice(["Pending", "Delivered", "Cancelled"])
        self.total_amount = round(random.uniform(10, 100), 2)
        self.payment_mode = random.choice(["Credit Card", "Cash", "UPI"])
        self.discount_applied = round(random.uniform(0, self.total_amount * 0.2), 2)
        self.feedback_rating = round(random.uniform(1, 5), 1) if self.status == "Delivered" else None

        return self


orders_list = []
num_orders = 100


for _ in range(num_orders):
    order = Order(None,None,None,None,None,None,None,None,None,None)
    orders_list.append(order.generate_random_order())


for order in orders_list:
    print(order)
```

```python
class zomato:
  def __init__(self,host,user,port,password,database):
    self.host = host
    self.user = user
    self.port = port
    self.password = password
    self.database = database
    self.conn = None
    self.cursor = None
    self.connect()
  def connect(self):
    self.conn = mysql.connector.connect(host = self.host, user = self.user, port = self.port, password = self.password, database = self.database)
    self.cursor = self.conn.cursor()
    print("connection successful")
  def create_table(self,table_name,columns):
    self.cursor.execute(f"CREATE TABLE IF NOT EXISTS {table_name} ({columns})")
    self.conn.commit()
    print("table created successfully")
  def insert_data(self,table_name,data):
    self.cursor.execute(f"INSERT INTO {table_name} VALUES {data}")
    self.conn.commit()
    print("data inserted successfully")
```
